# Задача 1

Города A, B, C вместе с соединяющими их прямыми дорогами образуют треугольник.
Известно, что прямой путь из А в B на 200 км короче, чем объезд через C, а прямой
путь из A в C на 300 км короче, чем объезд через B. Найдите расстояние между
городами B и С


## Решение 1:

Из условия получаем, что сумма двух путей (AB и BC) на 500 км меньше, чем сумма этих путей и дважды расстояние между городами B и С. Таким образом, расстояние между B и С есть половина 500 км, то есть 250 км.


## Ответ: 
Расстояние между городами B и С равно 250 км

## Решение 2:

Изобразим условние задачи:

<img src="img/three_cities.png" alt="три города" style="width: 400px;"/>

Получаем систему уравнений: 
$$
\begin{equation}
\left\{ \begin{array}{ll}
                c + 200 = b + a \\
                b + 300 = c + a
        \end{array}\right.
\Leftrightarrow       
\left\{ \begin{array}{ll}
                c + 200 - b=  a \\
                b + 300 - c=  a
        \end{array}\right.
\end{equation}
$$
То есть $2a = 500$, откуда $a = 250$

## Ответ: 
Расстояние между городами B и С равно 250 км

**При решении задачи полагалось, что** пути из X в Y, из Y в X и расстояние между готорадами X и Y совпадают. 

# Задача 2

Тестируется доставка последней мили с помощью автономных
роботов-доставщиков. 

**Алгоритм работы робота:** 
1. На сортировочном центре в черте города роботу загружают посылку (помещается только одна).
2. Робот отправляется по адресу, где его встречает получатель и робот отдает посылку.
3. Для доставки следующей посылки робот возвращается на сортировочный центр.

Для простоты расчетов **приняты допущения:**
- Сортировочный центр обслуживает район города, очерченный окружностью радиуса R.
- Клиенты по этой площади распределены равномерно.
- Робот двигается к клиенту вдоль радиуса.
- Скорость движения робота постоянна = v.
- Временем, необходимым на загрузку и разгрузку посылки, можно пренебречь (оно равно нулю).

**Вопрос**:
Как связано количество посылок, которое успеет доставить робот за день и
предельная дальность доставки R?


## Решение:

Допустим, что роботу поступает список заказов и пока эти заказы не будут выполенны список заказов не будет обновлен.
Определим **время на выполнение всех заказов** из этого списка.

**Плотность**

Пусть $\rho(x,y) = \rho$ - плотность заказов, то есть количество заказов на единицу площади. В данной постановке задачи плотность есть постоянная величина. Число заказов на некоторой площади $S$, определяется как 
$$ S \cdot \rho $$

**Система координат**

Время на доставку товаров зависит от расстояния заказчика до сортировочного центра (СЦ), то есть робот затратит одинаковое время на доставку всем заказчикам находящимся на одинаковом расстоянии до СЦ. Плотность постоянная, граница области работы СЦ $R$, поэтому будем рассматривать зависимость времени доставки только от $r$ (расстояния до СЦ). 

**Время на выполнение всех заказов**

Определим время доставки как сумму времен доставки на касающихся друг друга кольцах, покрывающих область доставки.

Для кольца изображенного на рисунке справа полагаем, что робот должен пролетать расстояние 
$$L = 2  \cdot (r+\Delta r/2)$$
Он затратит $L/v$ времени на этот полет. Всего таких полетов для данного кольца он совершит $ \Delta S \cdot \rho $. Таким образом, для доставки всех заказов для этого кольца уйдет времени: 

$$ \Delta T = \frac{\Delta S \cdot \rho \cdot 2  \cdot (r+\Delta r/2)}{v} =  \frac{ 2 \rho }{v} \cdot \Delta S \cdot (r+\Delta r/2)$$

где $\Delta S = \pi ((r+\Delta r)^{2} - r^2 ) = \pi (2 r \Delta r+ \Delta r^2 ) $.

Так как мы будем рассматривать поведение этой функции (времени доставки) при  $ \Delta r \rightarrow 0 $, то избавимся от слагаемых с $\Delta r$ в более чем первой степени. Получаем, что 

$$ \Delta T = \frac{ 2 \rho \pi }{v} \cdot  (2 r \Delta r ) \cdot (r+\Delta r/2) = \frac{ 4 \rho \pi }{v} \cdot  (r^2 \Delta r ) $$

Находим сумму для всех колец при $ \Delta r \rightarrow 0 $. Так как $\rho$ и $v$ не зависят от $r$, то запишем эту сумму как следующий интеграл.

$$ T =  \frac{ 4 \rho \pi }{v} \int_{0}^{R} (r^2 ) \,dr   $$

 По формуле Ньютона–Лейбница получаем:
 

$$ T =  \frac{ 4 \rho \pi R^3 }{3 v}   $$

Другими словами, для района радиуса $R$ доставка всех заказов в количестве 
$$ N =  S \cdot \rho  = \pi R ^2 \rho $$

потребует времени 
$$ T =  \frac{ 4 \rho \pi R^3 }{3 v} $$

То есть в среднем на доставку одного товара уйдет $ t_{avg} = \frac{T}{N} = \frac{4R}{3v} $ единиц времени.

**Количество посылок в отведенное время**

Исходим из того, что робот будет сначала доставлять товар самым близким заказчикам, так как в этом случа будет выполнено наибольшее число заказов. 

Тогда за отведенное время $t$ робот выполнял вылеты и максимальный радиус $r_{max}$ на котором он остановился определяется из соотношения
$$ t =  \frac{ 4 \rho \pi r_{max}^3 }{3 v} $$

Тогда 
$$ r_{max} =  \sqrt[3]{\frac{ 3 t v }{4 \rho \pi}} $$

**Допущение:** Для упрощения пусть робот успел посетить всех заказчиков с радиусом $r_{max}$

Получаем, что число выполененых заказов равно
$$ N = S \cdot \rho = \pi r_{max}^2 \cdot \rho = 
\frac{1}{2} 
\left( \frac{ \pi \rho }{2} \right)^{\frac{1}{3}} 
 \left(  3 t v \right)^{\frac{2}{3}}   $$ 

**В задаче стоял вопрос:**

Как связано количество посылок, которое успеет доставить робот за день и предельная дальность доставки R?


## Ответ: 

За отведенное время $t$, робот сумеет выполнить заказов 

$$ N =  \pi r_{max}^2 \cdot \rho = \frac{1}{2} 
\left( \frac{ \pi \rho }{2} \right)^{\frac{1}{3}} 
 \left(  3 t v \right)^{\frac{2}{3}}   ,$$

причем заказы будут доставляться не дальше чем: 
$$ r_{max} =  \sqrt[3]{\frac{ 3 t v }{4 \rho \pi}} ,$$

Здесь $v$ - скорость перевозки заказа,  $\rho$ - постоянная плотность заказов вокруг сортировочного центра.
